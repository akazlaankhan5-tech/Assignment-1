```python
marks = int(input("Enter your marks: "))

if marks < 25:
    print("Grade: F")
elif marks <= 45:
    print("Grade: E")
elif marks <= 50:
    print("Grade: D")
elif marks <= 60:
    print("Grade: C")
elif marks <= 80:
    print("Grade: B")
else:
    print("Grade: A")
```
